```python
#importing Libraries
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('DatasetM.xlsx')
df1.head()
```

```python
df2 = pd.read_excel('DatasetO.xlsx')
df2.head()
```

```python
## To make a new feature in 2nd Dataframe
df2['medicineType']= df2.medName.str.split('\s+').str[-1]
df2.head()
```

```python
##Cleaning df2
df2['medicin']= df2.medName.str.split('\s+').str[:-1]
df2['medicin']= df2.medicin.str.join(' ')

```

```python
## indexing of column
cols = df2.columns.tolist()
cols = cols[-1:] + cols[:-1]
df2 = df2[cols]
```

```python
df2.head()
```

```python
##drop extra medicin column
df2['medName'] = df2['medicin']
df2 = df2.drop(['medicin'],axis = 1)
cols = df2.columns.tolist()
cols = cols[-1:] + cols[:-1]
df2 = df2[cols]
df2.head()
```

```python
##Checking the medicinetype in df2
df2.medicineType.unique()
```

```python
##medicine type in df1
df1.medicineType.unique()
```

```python
### Now to match with medicine type we have to keep only those in df2 which are in df1
df2['medicineType'] = df2.medicineType.str.upper()
##we take a list in which all df1medicinetype are there 
med = ['TABLET', 'INJECTION', 'GEL', 'GENERAL', 'SYRUP', 'CAPSULE',
       'ROTACAP', 'RESPULES', 'ORAL SOLUTION', 'DRY SYRUP', 'POWDER',
       'LIQUID', 'SOLUTION', 'CLEANSER']
# we keep only those which medicine type are in the above list
df2 = df2[df2.medicineType.isin(med)]
df2.head()
```

```python
## Now we will drop some extra columns and keep those on which basis we will match
df1 = df1.drop(['medlife_price','manufacturer','Pack Size','pack_form'],axis =1)
df2 = df2.drop(['onemg_price','manufacturer','Pack Size','Pack Form','Units in Pack','Prescription Required','Stock'],axis = 1)

```

```python
df1.head()
```

```python
df2.head()
```

```python
df2 = df2[['medName','Ingredients','medicineType','Strength']]
df1.rename(columns={'strength':'Strength'}, inplace=True)
## do all in uppercase 
df2['medName'] = df2['medName'].str.upper()
df2['Ingredients'] = df2['Ingredients'].str.upper()
```

```python
##reset index of rows
df2 = df2.reset_index(drop = True)
## To extract Strength from ingridients
df2[['Ingredients','Strength']] = df2.Ingredients.str.split(' \(|\)', expand=True).iloc[:,[0,1]]

```

```python
df2.head()
```

```python
##Extract exact name from medName and remove strength in df2
df2['medicinename']= df2.medName.str.split('\s\d+').str[:-1]
df2['medicinename'] = df2['medicinename'].agg(lambda x: ''.join(map(str, x)))

df2.head()
```

```python
##Extract exact name from medName and remove strength in df1

df1['medicinename']= df1.medName.str.split('\s\d+').str[:-1]
df1['medicinename'] = df1['medicinename'].agg(lambda x: ''.join(map(str, x)))
df1.head()
```

```python
##Dropping the meName column
df1 = df1.drop(['medName'],axis = 1)
df2 = df2.drop(['medName'],axis = 1)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['Strength'] = df1['Strength'].str.replace(' ', '')
```

```python
 ##Just checking the Matching Name
set(df2['medicinename']).intersection(set(df1['medicinename']))
```

```python
##Matching ingredients name
set(df2['Ingredients']).intersection(set(df1['Ingredients']))
```

```python
intersected_df = pd.merge(df1, df2, how='inner')

print(intersected_df)
```

```python
df = intersected_df.drop_duplicates(keep=False) 
print(df)
```
